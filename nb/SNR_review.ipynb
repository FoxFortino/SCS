```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
results = pd.read_csv(
    "../data/snr_test/results.csv",
    names=[
        "noise_scale",
        "loss_trn",
        "ca_trn",
        "f1_trn",
        "loss_tst",
        "ca_tst",
        "f1_tst",
    ]
)
results = results[-101:]
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

ax.set_title("Injected Noise vs. Model Performance")
ax.scatter(results["noise_scale"], results["f1_tst"], color="k", marker="o", label="F1-Score (Tst)")
ax.scatter(results["noise_scale"], results["f1_trn"], color="k", marker="x", label="F1-Score (Trn)")

# ax.scatter(results["noise_scale"], results["ca_tst"], color="tab:blue", marker="o", label="CA (Tst)")
# ax.scatter(results["noise_scale"], results["ca_trn"], color="tab:blue", marker="x", label="CA (Trn)")

# ax.scatter(results["noise_scale"], results["loss_tst"], color="tab:orange", marker="*", label="Loss (Tst)")
# ax.scatter(results["noise_scale"], results["loss_trn"], color="tab:orange", marker="*", label="Loss (Trn)")

ax.set_xlabel("Scale of Injected Noise")
ax.set_ylabel("Model Metric")
ax.grid()
ax.legend(loc="upper right")
fig.show()
```
